```python
# import required libraries
from pywaffle import Waffle
import matplotlib.pyplot as plt
```

```python
# create a waffle chart
data = {'Data 1': 30, 'Data 2': 20, 'Data 3': 15, 'Data 4': 35}
fig = plt.figure(
    FigureClass=Waffle,
    rows=10,
    columns=10,
    values=data,
    legend={'loc': 'upper left', 'bbox_to_anchor': (1.1, 1)},
    icons='circle',
    icon_size=12,
    icon_legend=True)

# set chart attributes
fig.set_facecolor('#EAEAF2')
plt.title('Waffle Chart Example')
plt.show()
```

```python
#26_C_Write a chart title in centre location with large font size.
```

```python
# Create data
data = {'A': 20, 'B': 30, 'C': 50}

# Create a waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=10,
    columns=10,
    values=data,
    legend={'loc': 'upper left', 'bbox_to_anchor': (1.1, 1)},
    title={'label': 'Waffle Chart', 'loc': 'center', 'size': 15}
)
# Show the chart
plt.show()
```

```python
#26_D_Customize the spacing between tiles and aspect ratio.
```

```python
# Create data
data = {'A': 20, 'B': 30, 'C': 50}

# Create a waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=10,
    columns=10,
    values=data,
    legend={'loc': 'upper left', 'bbox_to_anchor': (1.1, 1)},
    interval_ratio_x=1,
    interval_ratio_y=2,
    block_aspect_ratio=1.25)
# Show the chart
plt.show()
```

```python
# Create data
data = {'A': 20, 'B': 30, 'C': 50}

# Create a waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=10,
    columns=10,
    values=data,
    vertical=True)
# Show the chart
plt.show()
```

```python
#26_G_Represent tiles with an array of color values
```

```python
# Create data
data = {'A': 20, 'B': 30, 'C': 50}
color=["#3366AA","#FFAA88","#EE1199"]

# Create a waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=10,
    columns=10,
    values=data,
    colors=color)
# Show the chart
plt.show()
```

```python
#26_H_Use cmap_name to specify a predefined color palette
```

```python
# Create data
data = {'A': 20, 'B': 30, 'C': 50}
 
# Create a waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=10,
    columns=10,
    values=data,
    cmap_name='Dark2')
# Show the chart
plt.show()
```
