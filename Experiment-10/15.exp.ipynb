```python
!pip install geopandas
!pip install descartes
!pip install matplotlib

# Import necessary libraries
import geopandas as gpd
import matplotlib.pyplot as plt

# Load world, Australia, and USA map data
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
aus = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
usa = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Subset Australia and USA data using their respective iso_a3 codes
aus = aus.loc[aus['iso_a3'] == 'AUS']
usa = usa.loc[usa['iso_a3'] == 'USA']

# Create a new figure and axis object
fig, ax = plt.subplots(figsize=(12, 6))

# Plot each map on the same axis object
world.plot(ax=ax, color='white', edgecolor='black')
aus.plot(ax=ax, color='green', edgecolor='black')
usa.plot(ax=ax, color='blue', edgecolor='black')

# Set the title of the plot
ax.set_title('World, Australia, and USA Maps')

# Show the plot
plt.show()
```

```python
why descartes library is used
ChatGPT
The Descartes library is used in conjunction with
the Matplotlib library to plot shapes on top of maps created with
the GeoPandas library. Specifically, Descartes provides a way to create matplotlib patches
that can be added to a GeoDataFrame plot. It is a small library that provides functionality 
to convert the geometric shapes in a GeoDataFrame into a format that Matplotlib can understand 
and then plot them on top of a map.
```
