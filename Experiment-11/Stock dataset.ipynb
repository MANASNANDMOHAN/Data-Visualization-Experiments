```python
#Using Stock dataset, display time series data of the all the attributes of the entire dataset (Open, Low, High, Last, Close, Total trade Quantity and Total Turnover (in Lacs.) in the same plot. Visualize the output and compare it.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the stock dataset
df = pd.read_csv('WIPRO.csv')

# Convert date column to datetime format and set as index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Plot time series data for all attributes
plt.figure(figsize=(15, 8))
plt.plot(df['Open'], label='Open')
plt.plot(df['Low'], label='Low')
plt.plot(df['High'], label='High')
plt.plot(df['Last'], label='Last')
plt.plot(df['Close'], label='Close')
plt.legend()
plt.title('Stock Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()
```
