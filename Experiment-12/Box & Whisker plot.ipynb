```python
import matplotlib.pyplot as plt
import numpy as np
 
np.random.seed(10)
dataset1=np.random.normal(100,10,220)
dataset2=np.random.normal(80,20,200)
dataset3=np.random.normal(70,30,150)
dataset4=np.random.normal(50,40,200)
dataset=[dataset1,dataset2,dataset3,dataset4]
figure=plt.figure(figsize=(10,8))
ax=figure.add_axes([0,0,1,1])
bp=ax.boxplot(dataset,notch='TRUE',vert=1)

plt.show()
```

```python
#############################################################################################
In box and whisker plots, hue is a parameter that allows us to divide the data into subgroups
and display them separately. 
By using hue, we can compare the distribution of a variable across different subgroups.

For example, if we have a box and whisker plot showing the distribution of age
for passengers on a ship, we can use hue to divide the passengers into different classes 
and show the distribution of age for each class separately. 
This can help us understand if there are any differences in the age distribution between 
the different classes of passengers.
#############################################################################################
```

```python
sns.boxplot(x="Pclass", y="Age", hue="Embarked", data=df)
```
