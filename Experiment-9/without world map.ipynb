```python
# Import necessary libraries
import geopandas as gpd
import matplotlib.pyplot as plt 
```

```python
# Load world map data
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Define the region to remove by selecting the row that matches the name of the region
region_to_remove = 'Greenland'
world_no_region = world[world.name != region_to_remove]

# Plot the world map without the selected region
fig, ax = plt.subplots(figsize=(10, 5))
world_no_region.plot(ax=ax)
ax.set_title('World Map without ' + region_to_remove)
plt.show()
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
fig, ax = plt.subplots(figsize=(10, 5))
world.plot(ax=ax)
ax.set_title('World Map with Greenland')
plt.show()
```
