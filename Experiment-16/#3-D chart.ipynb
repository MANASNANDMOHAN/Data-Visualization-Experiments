```python
#Write a program in python to draw a 3D line chart using the following:
a)Z axis - numbers between 0 to 15
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Generate some sample data
x = np.arange(0, 10, 1)
y = np.sin(x)
z = np.linspace(0, 15, len(x))

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the line chart
ax.plot(x, y, z)

# Set the axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
# Show the plot
plt.show()
```

```python
# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate data for the z-axis
z = np.linspace(0, 15, 100)

# Compute the sine of the z-axis values
x = np.sin(z)

# Plot the 3D line chart
ax.plot(z, x, zs=0, zdir='y')

# Set the labels and title
ax.set_xlabel('Z-axis')
ax.set_ylabel('X-axis')
ax.set_zlabel('Y-axis')
ax.set_title('3D Line Chart')
# Show the plot
plt.show()
```

```python
# Create data for the z-axis
z = np.linspace(0, 15, 100)

# Create data for the x-axis
x = np.sin(z)

# Create data for the y-axis
y = np.cos(z)

# Create a figure and a 3D axes object
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the line
ax.plot(x, y, z)

# Set labels for each axis
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis') 
# Show the plot
plt.show()
```
