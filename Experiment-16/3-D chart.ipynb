```python
Write a program in python to draw a 3D line chart and 3D scatterplot in the same space.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Generate some data for the 3D line chart and scatter plot
z = np.linspace(0, 15, 100)
x = np.sin(z)
y = np.cos(z)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D line chart
ax.plot(x, y, z, label='line chart')

# Generate some data for the 3D scatterplot
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
z_scatter = np.random.rand(50)

# Plot the 3D scatterplot
ax.scatter(x_scatter, y_scatter, z_scatter, c='r', label='scatter plot')

# Add labels and title to the plot
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Line Chart and Scatter Plot')
# Add legend to the plot
ax.legend()
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Generate data
z=np.linspace(0,2,1000)
x=z*np.sin(40*z)
y=z*np.cos(40*z)
# Create figure and 3D axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the line chart
ax.plot(x, y, z)

# Set the view angle to (30, 60) degrees
ax.view_init(elev=90, azim=60)

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Line Chart')

# Show the plot
plt.show()
```
