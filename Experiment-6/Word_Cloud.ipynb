```python
import markovify
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load the source text
with open("tushar.txt") as f:
    text = f.read()

# Build the Markov chain model
text_model = markovify.Text(text)

# Generate a new text of 400 characters
generated_text = text_model.make_short_sentence(400)

# Create and plot the word cloud
wordcloud = WordCloud().generate(generated_text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#Markovify
markovify is a Python library that provides a simple and efficient way to build and 
train Markov chain models for text generation.

Markov chain models are used to generate new sequences of text based on the patterns and structure of existing text. 
The basic idea is to use a statistical model to learn the probability distribution of the sequence of words 
in a given text corpus, and then use that model to generate new text that follows similar patterns and structure.

markovify simplifies the process of building and training Markov chain models for text generation. 
It provides an easy-to-use interface for building models of varying degrees of complexity 
and for generating new text based on those models.

Using markovify, you can:

Build a Markov chain model from a text corpus
Specify the order of the model (i.e., how many previous words to consider when predicting the next word)
Train the model on the text corpus
Generate new text that follows the same patterns and structure as the original text
Overall, markovify makes it easy to generate new text that is similar in style and structure to an existing text corpus,
which can be useful for a variety of applications such as chatbots, language generation,
and text summarization.
```

```python
#interpolation='bilinear'
interpolation is a parameter that can be specified when displaying 
or manipulating an image using a Python library such as Matplotlib or Pillow.

In image processing, interpolation refers to the technique of estimating the values of pixels 
at locations 
that are not explicitly defined in the image. This is often necessary when scaling or transforming an image,
or when zooming in on a particular region of the image.

The interpolation parameter allows you to specify the method used for 
interpolation. 'bilinear' is one of the most commonly used interpolation methods. 
Bilinear interpolation uses a weighted average of the four nearest 
neighboring pixels to estimate the value of the pixel 
at a new location. 
The resulting image is smoother and less pixelated than other interpolation methods 
such as nearest-neighbor interpolation, but it may also result in some loss of sharpness or detail.

When interpolation='bilinear', it means that the image will be displayed using bilinear interpolation.
This can be useful for creating a smoother, more visually appealing image, 
but it may also result in some loss of detail. 
The specific choice of interpolation method depends on the requirements of your application
and the characteristics of the image data.
```
