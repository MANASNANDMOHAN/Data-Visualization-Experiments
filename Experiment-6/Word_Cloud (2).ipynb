```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# Load the dataset
data = pd.read_csv("IMDB Dataset.csv")

# Define the custom stopword list
stopwords = set(STOPWORDS)
stopwords.update(["movie", "wonderful", "watch", "really", "good", "great","stunning"])

# Load the mask image and convert to numpy array
mask = np.array(Image.open("mask.png"))

# Define the word cloud object
wc = WordCloud(background_color="white", max_words=200, stopwords=stopwords, mask=mask, contour_width=3, contour_color='steelblue')

# Generate a word cloud
text = " ".join(review for review in data.review)
wc.generate(text)

# Plot the word cloud
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
######show mask image 2
```

```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
# Load the dataset
data = pd.read_csv("IMDB Dataset.csv")
data.head()
```

```python
# Define the custom stopword list
stopwords = set(STOPWORDS)
stopwords.update(["movie", "wonderful", "watch", "really", "good", "great","stunning"])

# Load the mask image and convert to numpy array
mask = np.array(Image.open("mask.png"))

# Define the word cloud object
wc = WordCloud(background_color="white", max_words=200, 
               stopwords=stopwords, mask=mask, contour_width=3, contour_color='steelblue')

# Generate a word cloud
text = " ".join(review for review in data.review)
wc.generate(text)

# Plot the word cloud
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#################################################################################################################
Masking is a technique used in word cloud creation to create a word cloud in a specific shape or image. 
When creating a word cloud with a mask, the words are placed only within the bounds of the mask image 
and are masked out everywhere else.

Masking is useful in creating custom-shaped word clouds that are more visually appealing 
and can convey a specific message or theme. 
For example, a word cloud about the ocean could be created with a mask in the shape of a wave or a fish, 
while a word cloud about a city could be created with a mask in the shape of a skyline or a landmark.

Masking can be done by loading an image as a numpy array 
and passing it to the WordCloud constructor using the mask parameter. 
The white pixels in the image represent the areas where words can be placed in the word cloud,
while the black pixels represent areas that will be masked out.

Overall, masking is a powerful tool that can enhance the visual impact of a word cloud 
and help to create a more engaging and memorable experience for the viewer.
################################################################################################################
```

```python
################################################################################################################
In natural language processing, stopwords are words that are commonly used in a language, 
but typically do not carry any significant meaning or contribute to the context of a sentence.
Examples of stopwords in English include "the", "and", "a", "an", "in", "of", and so on.

In word cloud creation, stopwords are often removed from the text before generating the word cloud. 
This is because including stopwords in the word cloud can result in the cloud being dominated by very common 
and unimportant words, which can make it harder to interpret or understand the main themes and topics in the text.

By removing stopwords, the word cloud can highlight the most relevant and important words in the text
, and make it easier to identify the key themes and topics.
################################################################################################################
```
